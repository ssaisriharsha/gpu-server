```python
from sentence_transformers import CrossEncoder
from datasets import load_dataset
import torch
from torch.nn.functional import softmax
import math
import pandas as pd
from huggingface_hub import login
```

```python
checkpoint="cross-encoder/stsb-roberta-large"
dataset="harsha2946/ParaDetox-Augmented"
ds=load_dataset(dataset)
```

```python
model=CrossEncoder(checkpoint, model_kwargs={"device_map": "auto"})
```

```python
def add_similarity_column(ex):
    similarity=model.predict([ex["en_toxic_comment"], ex["en_neutral_comment"]])
    return {
        "sentence_similarity": similarity
    }
```

```python
ds=ds.map(add_similarity_column, batched=False)
```

```python
df=pd.DataFrame(ds["train"])
df=df[df["sentence_similarity"]>=0.5]
df.describe()
```

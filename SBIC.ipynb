```python
from datasets import load_dataset
import html
import unicodedata
```

```python
ds=load_dataset("csv", data_files="project/SBIC/SBIC.v2.trn.csv", delimiter=",")
ds
```

```python
def keep_columns(ex_batch):
    return {
        "post": [unicodedata.normalize("NFKD", html.unescape(post)) for post in ex_batch["post"]],
        "offensive": [offensive for offensive in ex_batch["offensiveYN"]]
    }
```

```python
ds=ds.map(lambda ex_batch: keep_columns(ex_batch), batched=True, remove_columns=ds["train"].column_names)
```
